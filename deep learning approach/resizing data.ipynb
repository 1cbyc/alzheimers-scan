```python
output_dir = './resized_output'
os.makedirs(output_dir, exist_ok=True)

for files in tqdm(ls('slices')):
    print(files)
    folder_name = files.split('\\')[-2] if '\\' in files else files.split('/')[-2]
    file_name = files.split('\\')[-1] if '\\' in files else files.split('/')[-1]
    im = plt.imread(files)[5:265, 10:270, 0]
    im = resize(im, (260, 260), mode='constant')
    
    folder_path = os.path.join(output_dir, folder_name)
    os.makedirs(folder_path, exist_ok=True)
    output_path = os.path.join(folder_path, file_name + '.jpg')
    sio.imsave(output_path, im)
```
